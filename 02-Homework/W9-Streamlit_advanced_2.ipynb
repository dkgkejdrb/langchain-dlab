# 🧪 과제: 회원 가입 및 로그인 시스템 Streamlit으로 구현하기

## 🎯 과제 목표

**Streamlit을 활용하여** 간단한 **회원 가입 및 로그인 기능**을 구현해봅시다.  
회원 정보를 저장하고, 로그인 시 입력 검증 및 페이지 이동을 다루는 실습입니다.

## 📌 필수 구현 요소

다음 항목을 반드시 **Streamlit** 기반으로 구현해야 합니다:

- **회원 가입 페이지**
  - 사용자가 **아이디, 비밀번호, 나이, 학교**를 입력하여 가입할 수 있도록 구성
  - 회원 가입 성공 시, 딕셔너리에 정보를 저장
    ```python
    users = {"아이디": {"비밀번호": ..., "나이": ..., "학교": ...}}
    ```

- **로그인 페이지**
  - 사용자가 입력한 **아이디, 비밀번호**를 검증
  - 다음 상황에 따라 각각 다른 메시지를 표시:
    - **아이디 또는 비밀번호가 틀린 경우** ➔ 로그인 실패 메시지
    - **입력하지 않고 로그인 시도한 경우** ➔ 입력 누락 경고 메시지
    - **정확히 입력한 경우** ➔ 로그인 성공 메시지 표시 후 `main.py`로 이동 (`st.switch_page("main.py")` 활용)

- **페이지 이동**
  - 로그인 성공 시 `st.switch_page("main.py")`를 사용해 이동
